```python
# df['date_added'] = pd.to_datetime(df['date_added'])
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')

df['year_added'] = df['date_added'].dt.year
```

```python
# Fill missing values with 'Unknown' or a default
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unrated', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['date_added', 'year_added'])
```

```python
from collections import Counter

genres = df['listed_in'].dropna().str.split(', ')
flat_genres = [genre for sublist in genres for genre in sublist]
top_genres = pd.Series(Counter(flat_genres)).sort_values(ascending=False).head(10)

top_genres.plot(kind='bar', color='purple')
plt.title("Top 10 Most Common Genres on Netflix")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
genre_counts = pd.Series(Counter(flat_genres)).sort_values(ascending=False)
genre_counts.head(10)
```

```python
top_directors = df['director'].value_counts().head(10)

top_directors.plot(kind='barh', color='indigo')
plt.title("Top 10 Most Frequent Directors on Netflix")
plt.xlabel("Number of Titles")
plt.gca().invert_yaxis()
plt.show()

```

```python
cast_list = df['cast'].dropna().str.split(', ')
flat_cast = [actor for sublist in cast_list for actor in sublist]
top_actors = pd.Series(Counter(flat_cast)).sort_values(ascending=False).head(10)

top_actors.plot(kind='bar', color='darkgreen')
plt.title("Top 10 Most Featured Actors on Netflix")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

```python
df[['duration_value', 'duration_unit']] = df['duration'].str.extract(r'(\d+)\s+(\w+)')
df['duration_value'] = pd.to_numeric(df['duration_value'], errors='coerce')
```

```python
movie_durations = df[df['type'] == 'Movie']['duration_value'].value_counts().head(10)

movie_durations.plot(kind='bar', color='tomato')
plt.title("Top 10 Most Common Movie Durations (minutes)")
plt.xlabel("Duration (minutes)")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
df['month_added'] = df['date_added'].dt.month_name()

month_counts = df['month_added'].value_counts().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])

month_counts.plot(kind='bar', color='steelblue', figsize=(10,5))
plt.title("Netflix Titles Added by Month")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
df['month_added'] = df['date_added'].dt.month_name()

month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

df['month_added'] = pd.Categorical(df['month_added'], categories=month_order, ordered=True)

df['month_added'].value_counts().sort_index().plot(kind='bar', color='coral', figsize=(10,5))
plt.title("Titles Added by Month")
plt.xlabel("Month")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
type_year = df.groupby(['year_added', 'type']).size().unstack().fillna(0)

type_year.plot(kind='line', figsize=(10,5), marker='o')
plt.title("Movies vs TV Shows Added Per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
# Create a cross-tab of country vs genre
genre_country = df[['country', 'listed_in']].dropna()

# We'll just keep the top 5 countries for simplicity
top_countries = df['country'].value_counts().head(5).index.tolist()

filtered = genre_country[genre_country['country'].isin(top_countries)]

#connting how often genares are appearing 
from collections import defaultdict

country_genre_count = defaultdict(Counter)

for _, row in filtered.iterrows():
    country = row['country']
    genres = row['listed_in'].split(', ')
    for g in genres:
        country_genre_count[country][g] += 1

genre_df = pd.DataFrame(country_genre_count).fillna(0).astype(int)

# Heatmap
plt.figure(figsize=(12,6))
sns.heatmap(genre_df, cmap='YlGnBu', annot=True)
plt.title("Genre Popularity by Country (Top 5)")
plt.show()
```
